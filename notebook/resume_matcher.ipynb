```python
import re, string, PyPDF2, spacy
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
nlp = spacy.load('en_core_web_sm')
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if token.text not in stop_words]
    return ' '.join(tokens)
```

```python
def calculate_similarity(resume, job):
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([resume, job])
    return cosine_similarity(vectors[0], vectors[1])[0][0]

resume = 'Python developer with machine learning experience'
job = 'Looking for Python engineer with ML skills'
print('Match Score:', round(calculate_similarity(preprocess_text(resume), preprocess_text(job))*100, 2), '%')
```
